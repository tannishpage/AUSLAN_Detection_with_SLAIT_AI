# Recurrence Plots in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Making Time Series Data 

```python
## Task 1a Create a Time series
nums = np.arange(1, 200)
time_series = []
for num in nums:
    a = np.math.sin(num*2*np.math.pi*(8/1000))
    time_series.append(a)
plt.plot(nums, time_series)
plt.show()
```

```python
## Task 1b Try different frequencies

nums = np.arange(1, 200)
time_series = []
frequency = 23/1000
for num in nums:
    a = np.math.sin(num*2*np.math.pi*(frequency))
    time_series.append(a)
plt.plot(nums, time_series)
plt.show()
```

```python
## Task 1c Add 2 frequencies and plot

nums = np.arange(1, 200)
time_series = []
frequency1 = 8/1000
frequency2 = 23/1000
for num in nums:
    a = np.math.sin(num*2*np.math.pi*(frequency1))
    b = np.math.sin(num*2*np.math.pi*(frequency2))
    time_series.append(0.6*a + 0.4*b)
plt.plot(nums, time_series)
plt.show()
```

### Recurrence Plots

The `dist` function from matlab as used in the PDF

```python
def calculate_dist(pos, all_pos):
    distances = []
    for pos2 in all_pos:
        distance = np.math.sqrt((pos - pos2)**2)
        distances.append(distance)
    return np.array(distances)
```

```python
## Task 2a Create a distance plot
distance_matrix = []
for x in time_series:
    distance_matrix.append(calculate_dist(x, time_series))
distance_matrix = 1 - np.array(distance_matrix)
print(np.where((distance_matrix) > 0.9))
im = plt.imshow((distance_matrix))
cbar = plt.colorbar(im)
cbar.set_label("Colorbar")
plt.show()
```

```python
## Task 2b Create a recurrence plot with threshold 0.9

distance_matrix = []
for x in time_series:
    distance_matrix.append(calculate_dist(x, time_series))
distance_matrix = 1 - np.array(distance_matrix)
print(np.where((distance_matrix) > 0.9))
distance_matrix[np.where((distance_matrix) <= 0.9)] = 0
distance_matrix[np.where((distance_matrix) > 0.9)] = 1
im = plt.imshow((distance_matrix), cmap='gray')
cbar = plt.colorbar(im)
cbar.set_label("Colorbar")
plt.show()
```

```python
## Task 2c Vary the threshold

distance_matrix = []
for x in time_series:
    distance_matrix.append(calculate_dist(x, time_series))
distance_matrix = 1 - np.array(distance_matrix)
print(np.where((distance_matrix) > 0.9))
distance_matrix[np.where((distance_matrix) <= 0.5)] = 0
distance_matrix[np.where((distance_matrix) > 0.5)] = 1
im = plt.imshow((distance_matrix), cmap='gray')
cbar = plt.colorbar(im)
cbar.set_label("Colorbar")
plt.show()
```

```python
## Task 2d Calculate the average recurrence

distance_matrix = []
for x in time_series:
    distance_matrix.append(calculate_dist(x, time_series))
distance_matrix = 1 - np.array(distance_matrix)
print(np.where((distance_matrix) > 0.9))
average_recurrence = np.sum(np.sum(distance_matrix[np.where((distance_matrix) > 0.9)]))/(200*200)
distance_matrix[np.where((distance_matrix) <= 0.9)] = 0
distance_matrix[np.where((distance_matrix) > 0.9)] = 1
im = plt.imshow((distance_matrix), cmap='gray')
cbar = plt.colorbar(im)
cbar.set_label("Colorbar")
plt.show()
print("Average Recurrence:", average_recurrence)
```
